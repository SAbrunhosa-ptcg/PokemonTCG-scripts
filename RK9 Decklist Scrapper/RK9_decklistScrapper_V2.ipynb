```python
### Import saved data ###
filename = "Tournament data - day 1/ImportedData_EUIC2025.xlsx"
# there are some complications with importing since all imports are strings and some columns contain other things 
# should clean this up later, but for now the decklist is all that matters
player_list = f.ImportData(filename) 

#print(np.array(eval(player_list[0].decklist)).tolist())

### Complete information ###
#day2_limit = 184 # Merida
day2_limit = 361 # EUIC

for player in player_list:
    # CP and Day 2
    player.calculateCP()
    player.checkDay2(day2_limit)

    # pokemon list
    player.pokemonlist = f.GetPokemonList(player.decklist)

    # archetype and variant
    player.archetype, player.variant = f.GetArchetype(player.pokemonlist) # variant is not implemented for now

print(player_list[0].CP)
print(player_list[0].decklist)
print(player_list[0].pokemonlist)
print(player_list[0].archetype)
print(player_list[0].variant)

### Save complete data ###
f.ExportData(player_list, filename) # Export to excel
```

```python
# loop to count individual cards
for player in player_list:
    if player.archetype == 'Lugia (Cinccino)':
        #print(player.decklist)
        lumineon_count = 0
        for card in player.decklist:
            #print(card[1])
            if card[1] == 'Lumineon V':
                lumineon_count += int(card[2])
        if lumineon_count != 2:
            print(lumineon_count)
            print(player.name)
            print(player.decklist_link)
```
